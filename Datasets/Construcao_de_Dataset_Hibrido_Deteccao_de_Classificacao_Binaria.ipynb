### 2.2&nbsp;Redimensionamento de Arquivo


* `Compatibilidade com Arquiteturas de Modelo:` Algumas arquiteturas de rede neural, especialmente as pré-treinadas, exigem que as imagens de entrada tenham dimensões específicas. O redimensionamento de imagem permite ajustar o conjunto de dados às exigências dessas arquiteturas, facilitando a utilização de redes poderosas e bem-sucedidas em suas tarefas.

* `Impacto no Desempenho do Modelo:` Diferentes tamanhos de imagem podem afetar a precisão do modelo. Imagens muito pequenas podem não ter detalhes suficientes para uma classificação precisa, enquanto imagens muito grandes podem incluir ruído desnecessário. O redimensionamento pode ajuda a determinar o tamanho ideal para a melhor performance do modelo.

* `Otimização de Recursos Computacionais:` Imagens grandes podem consumir significativamente mais memória e poder de processamento. Ao entender as dimensões das imagens, você pode redimensioná-las para um tamanho que equilibre a qualidade necessária para uma boa detecção/classificação e a eficiência computacional, evitando o uso excessivo de recursos.

### 2.3&nbsp;Análise de Formato e Conversão de Arquivo


* `Análise de Formato de Arquivo:` A análise de formato de arquivo envolve identificar e compreender o tipo de arquivo de cada imagem (por exemplo, JPG, PNG, JPEG, TIFF, etc.). Isso é importante porque diferentes formatos podem ter diferentes propriedades, como suporte a transparência ou compressão com perdas, que podem afetar como as imagens são processadas e usadas no problema que esteja sendo trabalhado.

* `Conversão de Arquivo:` Depois de entender os formatos de suas imagens, o próximo passo é convertê-las para o formato desejado. A conversão de arquivo é o processo de mudar o formato de uma imagem para outro, geralmente para padronização ou compatibilidade com o problema e modelo específico.

### 2.4&nbsp;Padronização de Renomeamento Carvalheiro - O&S

**Renomeamento Original & Sintético**

> A **padronização de renomeamento Carvalheiro**, `Classe_SDSn°`, é fundamental para **distinguir os dados** com os quais você está trabalhando e, principalmente, para a **manutenção futura desse dataset**, pois assim você consegue identificar qual é o **status do conjunto de dados e quantificar as imagens** com as quais você está trabalhando através de uma **sistematização**.



* `Classe:` é a categoria da imagem, por exemplo, "Murilo" para imagens de uma pessoa chamada Murilo ou "Outros" para categorias diferentes.

* `Status do Dataset:` é um identificador, como "O" para Original ou "S" para Sintético.

* `n°:` é um número sequencial que começa em 1 e aumenta para cada imagem no diretório, garantindo que cada imagem tenha um identificador único.

### 2.5&nbsp;Técnica de Argumentação

**Desafio de Dados Limitados**
> Enfrentamos  um desafio devido ao número limitado de imagens do Murilo (fotos com status original, 44 imagens);

> Em aprendizado de máquina, um pequeno conjunto de dados pode limitar seriamente a capacidade do modelo de aprender variabilidade e generalizar bem para novos dados;

>Basicamente, se nosso modelo aprender apenas com poucas imagens, ele pode não reconhecer o Murilo em situações ligeiramente diferentes das fotos de treinamento.

**Implementação da Técnica de Augmentation**

* Para superar a limitação de dados, nosso grupo implementou a técnica de argumentação.

* Isso envolveu a criação de novas imagens através da alteração das originais de maneiras específicas, como girar, mudar a escala, cortar, ajustar a iluminação das imagens, etc.

* Essas alterações ajudam a simular como o Murilo poderia aparecer em diferentes cenários, condições de iluminação ou ângulos.
